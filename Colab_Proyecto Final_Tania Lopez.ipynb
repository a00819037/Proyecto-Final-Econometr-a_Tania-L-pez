```python
# =============================================================================
#                   COMPLETE INTRADAY ECONOMETRIC ANALYSIS
#             One single cell – separated by clear section titles
# =============================================================================

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.vector_ar.vecm import coint_johansen
import warnings

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
%matplotlib inline

# ──────────────── CONFIGURATION ────────────────
assets     = ['MSFT', 'JPM', 'CVX', 'CAT', 'KO', 'UNH', 'DIA']
timeframes = ['1M1D', '5M5D']           # corresponds to your file names
base_path  = '/content/drive/MyDrive/Colab Stocks Data/'

print("≡"*80)
print(" ANALYSIS START – All assets × both timeframes in one run ")
print("≡"*80)

# ──────────────── HELPER FUNCTIONS ────────────────
def load_series(asset, tf):
    path = f"{base_path}{asset}.{tf}.xlsx"
    try:
        df = pd.read_excel(path)
    except Exception as e:
        print(f"ERROR loading {asset}.{tf}: {e}")
        return None, None

    # Try to guess datetime and close columns
    dt_candidates = [c for c in df.columns if 'date' in c.lower() or 'time' in c.lower()]
    close_candidates = [c for c in df.columns if 'close' in c.lower()]

    dt_col  = dt_candidates[0] if dt_candidates else df.columns[0]
    close_col = close_candidates[0] if close_candidates else df.columns[-1]

    df[dt_col] = pd.to_datetime(df[dt_col], errors='coerce')
    df = df.dropna(subset=[dt_col]).set_index(dt_col)
    series = df[close_col].dropna().astype(float)

    print(f"  {asset}.{tf}  →  {len(series)} obs  |  index: {dt_col}  |  price: {close_col}")
    return series, df

# ──────────────── MAIN LOOP – EVERY ASSET × EVERY TIMEFRAME ────────────────
stationary_dict = {}

for tf in timeframes:
    print(f"\n\n{'═'*80}")
    print(f"                    TIMEFRAME: {tf}   (1D/1min or 5D/5min)")
    print(f"{'═'*80}\n")

    for asset in assets:
        print(f"\n{'─'*70}")
        print(f" ASSET: {asset}  –  {tf}")
        print(f"{'─'*70}")

        series, _ = load_series(asset, tf)
        if series is None or len(series) < 20:
            print(" → Skipping – not enough data")
            continue

        # ──────────────── 1. UNIT ROOT & STATIONARITY TESTS ────────────────
        print("\n[1] UNIT ROOT TESTS")
        print("─"*60)

        # ADF
        adf_res = adfuller(series, autolag='AIC')
        adf_stat, adf_p = adf_res[0], adf_res[1]
        print(f"ADF:  stat = {adf_stat:.4f}   p-value = {adf_p:.4f}")
        print("   → Stationary" if adf_p < 0.05 else "   → Unit root (non-stationary)")

        # KPSS
        kpss_res = kpss(series, regression='c', nlags='auto')
        kpss_stat, kpss_p = kpss_res[0], kpss_res[1]
        print(f"KPSS: stat = {kpss_stat:.4f}   p-value = {kpss_p:.4f}")
        print("   → Stationary" if kpss_p > 0.05 else "   → Non-stationary")

        is_stationary = (adf_p < 0.05) and (kpss_p > 0.05)
        stationary_dict[(asset, tf)] = is_stationary
        print(f"Conclusion: {'STATIONARY' if is_stationary else 'NON-STATIONARY / I(1)'}")

        # ──────────────── 2. CORRELOGRAM (ACF + PACF) ────────────────
        print("\n[2] CORRELOGRAM")
        print("─"*60)
        fig, ax = plt.subplots(2, 1, figsize=(11, 7))
        plot_acf(series, lags=40, ax=ax[0], title=f'ACF – {asset} {tf}')
        plot_pacf(series, lags=40, ax=ax[1], title=f'PACF – {asset} {tf}', method='ywm')
        plt.tight_layout()
        plt.show()

        # ──────────────── 3. ARIMA MODEL (with automatic order search) ────────────────
        print("\n[3] ARIMA MODEL")
        print("─"*60)
        best_aic = np.inf
        best_order = None

        for p in range(0, 4):
            for d in range(0, 3):
                for q in range(0, 4):
                    try:
                        m = ARIMA(series, order=(p,d,q)).fit()
                        if m.aic < best_aic:
                            best_aic = m.aic
                            best_order = (p,d,q)
                    except:
                        pass

        if best_order:
            print(f"Best ARIMA{best_order}   AIC = {best_aic:.2f}")
            model_arima = ARIMA(series, order=best_order).fit()
            print(model_arima.summary().tables[1])  # coefficients table
        else:
            print("No valid ARIMA model found in search range.")

        # ──────────────── 4. ARMA MODEL (on levels or differences) ────────────────
        print("\n[4] ARMA MODEL")
        print("─"*60)

        if is_stationary:
            data_arma = series
            print("→ Using original levels (stationary)")
        else:
            data_arma = series.diff().dropna()
            print("→ Using first differences")

        best_aic_arma = np.inf
        best_order_arma = None

        for p in range(0, 5):
            for q in range(0, 5):
                try:
                    m = ARIMA(data_arma, order=(p,0,q)).fit()
                    if m.aic < best_aic_arma:
                        best_aic_arma = m.aic
                        best_order_arma = (p,q)
                except:
                    pass

        if best_order_arma:
            print(f"Best ARMA{best_order_arma}   AIC = {best_aic_arma:.2f}")
            model_arma = ARIMA(data_arma, order=(*best_order_arma, 0)).fit()
            print(model_arma.summary().tables[1])
        else:
            print("No valid ARMA model found.")

        print("\n")

# ──────────────── 5. COINTEGRATION ANALYSIS (per timeframe) ────────────────
print(f"\n\n{'═'*80}")
print("               COINTEGRATION ANALYSIS – both timeframes")
print(f"{'═'*80}\n")

for tf in timeframes:
    print(f"\n[5] COINTEGRATION – {tf}")
    print("─"*70)

    series_dict = {}
    for asset in assets:
        s, _ = load_series(asset, tf)
        if s is not None:
            series_dict[asset] = s

    if len(series_dict) < 2:
        print("Not enough series loaded.")
        continue

    df_multi = pd.DataFrame(series_dict).dropna(how='any')
    print(f"Aligned data shape: {df_multi.shape}")

    if len(df_multi) < 30:
        print("→ Too few aligned observations – skipping Johansen")
        continue

    # Johansen on all assets
    joh = coint_johansen(df_multi, det_order=0, k_ar_diff=2)
    trace_stats = joh.lr1
    crit_95 = joh.cvt[:,1]

    print("\nJohansen Trace Test (all assets)")
    print("───────────────────────────────")
    for r in range(len(trace_stats)):
        print(f"r ≤ {r-1 if r>0 else 0:<2}  trace = {trace_stats[r]:8.2f}   crit 95% = {crit_95[r]:6.2f}   {'reject' if trace_stats[r] > crit_95[r] else 'do not reject'}")

    rank = sum(trace_stats > crit_95)
    print(f"\nEstimated cointegration rank (95%): {rank}")

    # Some interesting pairs
    pairs = [('MSFT','JPM'), ('MSFT','CVX'), ('MSFT','CAT'), ('MSFT','KO'), ('MSFT','UNH'), ('MSFT','DIA'), ('JPM','CVX'), ('JPM','CAT'), ('JPM','KO'), ('JPM','UNH'), ('JPM','DIA'), ('CVX','CAT'), ('CVX','KO'), ('CVX','UNH'), ('CVX','DIA'), ('CAT','KO'), ('CAT','UNH'), ('CAT','DIA'), ('KO','UNH'), ('KO','DIA'), ('UNH','DIA')]
    print("\nSelected pairs:")
    for a,b in pairs:
        if a in df_multi and b in df_multi:
            pair_df = df_multi[[a,b]].dropna()
            if len(pair_df) > 20:
                j = coint_johansen(pair_df, det_order=0, k_ar_diff=2)
                rank_pair = sum(j.lr1 > j.cvt[:,1])
                print(f"  {a} – {b:6}   rank = {rank_pair}   {'cointegrated' if rank_pair >= 1 else ''}")

print("\n" + "═"*80)
print("ANALYSIS FINISHED")
print("Review the printed results and plots above.")
print("═"*80)
```
